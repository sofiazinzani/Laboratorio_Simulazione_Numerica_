## Codice

```python
fig, axes=plt.subplots(nrows=2, ncols=2, figsize=(20,15))
axes=axes.flatten()



points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
#_____________________


axes[0].set_ylabel('Internal energy/nspin')


e = -J*( th + ch*thN )/( 1 + thN )
axes[0].plot(T, e, label= "analytical")


file=open("ISING_1D_DataBlocking/Risultati/Metropolis/output.ene.final", "r")
x=np.loadtxt("ISING_1D_DataBlocking/Risultati/Metropolis/output.ene.final", usecols=(0))
f=np.loadtxt("ISING_1D_DataBlocking/Risultati/Metropolis/output.ene.final", usecols=(2))
error=np.loadtxt("ISING_1D_DataBlocking/Risultati/Metropolis/output.ene.final", usecols=(3))
axes[0].errorbar(x,f,yerr=error, color = 'g', elinewidth=1, capsize=4,marker='s',mfc='blue',mec='green', ms=5, mew=1, label="Metropolis")


file=open("ISING_1D_DataBlocking/Risultati/Gibbs/output.ene.final", "r")
x=np.loadtxt("ISING_1D_DataBlocking/Risultati/Gibbs/output.ene.final", usecols=(0))
f=np.loadtxt("ISING_1D_DataBlocking/Risultati/Gibbs/output.ene.final", usecols=(2))
error=np.loadtxt("ISING_1D_DataBlocking/Risultati/Gibbs/output.ene.final", usecols=(3))
axes[0].errorbar(x,f,yerr=error, color = 'm', elinewidth=1, capsize=4,marker='s',mfc='red',mec='green', ms=5, mew=1, label="Gibbs")



#______________ Capacity
axes[1].set_ylabel('Capacity')


heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

axes[1].plot(T, heat, label= "analytical")



file=open("ISING_1D_DataBlocking/Risultati/Metropolis/output.capacity.final", "r")
x=np.loadtxt("ISING_1D_DataBlocking/Risultati/Metropolis/output.capacity.final", usecols=(0))
f=np.loadtxt("ISING_1D_DataBlocking/Risultati/Metropolis/output.capacity.final", usecols=(2))
error=np.loadtxt("ISING_1D_DataBlocking/Risultati/Metropolis/output.capacity.final", usecols=(3))
axes[1].errorbar(x,f,yerr=error, color = 'g', capsize=4,marker='s',mfc='blue',mec='green', ms=5, mew=1, label="Metropolis")

file=open("ISING_1D_DataBlocking/Risultati/Gibbs/output.capacity.final", "r")
x=np.loadtxt("ISING_1D_DataBlocking/Risultati/Gibbs/output.capacity.final", usecols=(0))
f=np.loadtxt("ISING_1D_DataBlocking/Risultati/Gibbs/output.capacity.final", usecols=(2))
error=np.loadtxt("ISING_1D_DataBlocking/Risultati/Gibbs/output.capacity.final", usecols=(3))
axes[1].errorbar(x,f,yerr=error, color = 'm', capsize=4,marker='s',mfc='red',mec='green', ms=5, mew=1, label="Gibbs")






#_______ Susceptibility

axes[2].set_ylabel('Susceptibility')

X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
axes[2].plot(T, X, label= "analytical")


file=open("ISING_1D_DataBlocking/Risultati/Metropolis/output.susceptibility.final", "r")
x=np.loadtxt("ISING_1D_DataBlocking/Risultati/Metropolis/output.susceptibility.final", usecols=(0))
f=np.loadtxt("ISING_1D_DataBlocking/Risultati/Metropolis/output.susceptibility.final", usecols=(2))
error=np.loadtxt("ISING_1D_DataBlocking/Risultati/Metropolis/output.susceptibility.final", usecols=(3))
axes[2].errorbar(x,f,yerr=error, color = 'g', capsize=4,marker='s',mfc='blue',mec='green', ms=5, mew=1, label="Metropolis")



file=open("ISING_1D_DataBlocking/Risultati/Gibbs/output.susceptibility.final", "r")
x=np.loadtxt("ISING_1D_DataBlocking/Risultati/Gibbs/output.susceptibility.final", usecols=(0))
f=np.loadtxt("ISING_1D_DataBlocking/Risultati/Gibbs/output.susceptibility.final", usecols=(2))
error=np.loadtxt("ISING_1D_DataBlocking/Risultati/Gibbs/output.susceptibility.final", usecols=(3))
axes[2].errorbar(x,f,yerr=error, color = 'm', capsize=4,marker='s',mfc='red',mec='green', ms=5, mew=1, label="Gibbs")


#______________ Magnetization

axes[3].set_ylabel('Magnetization')

h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
axes[3].plot(T, M, label= "analytical")

file=open("ISING_1D_DataBlocking/Risultati/Metropolis/output.magnetization_h002.final", "r")
x=np.loadtxt("ISING_1D_DataBlocking/Risultati/Metropolis/output.magnetization_h002.final", usecols=(0))
f=np.loadtxt("ISING_1D_DataBlocking/Risultati/Metropolis/output.magnetization_h002.final", usecols=(2))
error=np.loadtxt("ISING_1D_DataBlocking/Risultati/Metropolis/output.magnetization_h002.final", usecols=(3))
axes[3].errorbar(x,f,yerr=error, color = 'g', capsize=4,marker='s',mfc='blue',mec='green', ms=5, mew=1, label="Metropolis")



file=open("ISING_1D_DataBlocking/Risultati/Gibbs/output.magnetization_h002.final", "r")
x=np.loadtxt("ISING_1D_DataBlocking/Risultati/Gibbs/output.magnetization_h002.final", usecols=(0))
f=np.loadtxt("ISING_1D_DataBlocking/Risultati/Gibbs/output.magnetization_h002.final", usecols=(2))
error=np.loadtxt("ISING_1D_DataBlocking/Risultati/Gibbs/output.magnetization_h002.final", usecols=(3))
axes[3].errorbar(x,f,yerr=error, color = 'm', capsize=4,marker='s',mfc='red',mec='green', ms=5, mew=1, label="Gibbs")



#_____________________________
for ax in axes:
    ax.set_title('Ising 1D')
    ax.set_xlabel('T')
    ax.grid(True)
    ax.legend()


    
fig.tight_layout()
plt.show()
```

Dai grafici osserviamo l'accordo molto stretto tra i dati ottenuti dalle simulazioni (con entrambi i metodi di campionamento) e le curve teoriche che si ottengono dalle soluzioni analitiche del modello di Ising 1D.
