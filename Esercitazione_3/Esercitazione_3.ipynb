# Esercizio 3


I grafici riportati mostrano l'andamento del prezzo di un obbligazione europea con opzione di tipo *Put* e *Call*. 
In particolare, nei primi due grafici il prezzo finale dell'asset viene direttamente calcolato al tempo T, mentre per la seconda coppia si è stimato l'andamento del prezzo a intervalli temporali intermedi $S(t_i)$.

Nel primo caso si utilizza la funzione *GBM_Direct* che restituisce: $$S(T) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)T+\sigma W(T)\right]$$ dove $W(T)$ è una variabile gaussiana $\mathcal{N}(0,t)$. 

Nel secondo caso si utilizza per stimare il prezzo dell'asset al passo temporale successivo: $$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    
    
Possiamo notare che i metodi sono equivalenti.
Osserviamo quindi che i dati ottenuti sono compatibili. 
Infine notiamo che essi sono anche compatibili con la soluzione analitica ottenuta da Black-Scholes, che fornisce:

call:  14.975790778311286

put:  5.4595325819072364




```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


fig=plt.figure(figsize=(16, 6))

plt.subplot(1,2,1)
x, f, error=np.loadtxt("3.1.1/Dati.dat", usecols=(0,1,2), delimiter='\t', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.hlines(
    14.975790778311286,
    0,
    len(x),
    colors = "red",
    linestyle = "dotted",
    label = "Analytical solution"
)

plt.xlabel('Numer of block')
plt.ylabel('Option price')
plt.title('Call Option')
plt.grid(True)



plt.subplot(1,2,2)

x, f, error=np.loadtxt("3.1.1/Dati.dat", usecols=(0,3,4), delimiter='\t', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.hlines(
    5.4595325819072364,
    0,
    len(x),
    colors = "red",
    linestyle = "dotted",
    label = "Analytical solution"
)

plt.xlabel('Numer of block')
plt.ylabel('Option price')
plt.title('Put Option')
plt.grid(True)

plt.show()
```
