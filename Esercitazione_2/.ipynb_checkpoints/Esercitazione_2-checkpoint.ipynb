# Esercizio 2.1

Calcolo il seguente integrale via Monte Carlo:
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$



## ESERCIZIO 2.1.1

### Campionamento uniforme tra [0,1]

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error=np.loadtxt("2.1/2.1.1/Dati.dat", usecols=(0,1,2), delimiter='\t', unpack='true')

plt.figure(figsize=(9,6))
plt.errorbar(x,f,yerr=error)

plt.hlines(
    1,
    0, # Start of horizontal line
    x[-1], # End == total number of throws, i.e. input data points.
    colors = "red",
    linestyle = "dotted",
    label = "Expected value"
)

plt.xlabel('x')
plt.ylabel('I(x)')
plt.grid(True)
plt.legend(['Valore atteso', 'Integrale calcolato'], loc='best')

plt.show()
```

### Importance Sampling

```python
x, f, error=np.loadtxt("2.1/2.1.2/Dati.dat", usecols=(0,1,2), delimiter='\t', unpack='true')

plt.figure(figsize=(9,6))
plt.errorbar(x,f,yerr=error)


plt.hlines(
    1,
    0, # Start of horizontal line
    x[-1], # End == total number of throws, i.e. input data points.
    colors = "red",
    linestyle = "dotted",
    label = "Expected value"
)

plt.legend(['Valore atteso', 'Integrale calcolato'], loc='best')
plt.xlabel('x')
plt.ylabel('I(x)')
plt.grid(True)

plt.ylim(0.9,1.1)

plt.show()
```

## Esercizio 2.2

Per risolvere questo esercizio si è generato un vettore lungo $Npassi= 100$, dato da $Nblocchi=10000$ realizzazioni di un Random Walk. A ogni step di ogni cammino viene sommata la distanza a cui si è arrivati rispetto all'origine e il suo quadrato.
Terminati tutti gli spostamenti si calcola per ogni step $i\in [0,10^2]$ la media su tutti i cammini, l'errore statistico di tale media.
Di seguito sono riportati i grafici per un RW generato su un reticolo quadrato e per punti distuibuiti in modo continuo.


### 2.2.1 Random walk su un reticolo quadrato, di passo 1.

```python
x, f, error=np.loadtxt("2.2/2.2.1/Dati.dat", usecols=(0,1,2), unpack='true')

plt.figure(figsize=(9,6))
plt.errorbar(x,f,yerr=error, linewidth=4, zorder=0)

plt.plot(x,pow(x, 0.5), color='r', linewidth=1.2, zorder=10)

plt.legend(['Np^(1/2)', 'Dati'], loc='best')
plt.xlabel('Np')
plt.ylabel('Radice quadrata della distanza quadratica da origine')
plt.grid(True)

plt.show()
```

### 2.2.2 Random walk nel continuo, ottenuto con uno step di lunghezza 1, lungo una direzione casuale, ottenuta estraendo uniformemente $\theta$ tra [$0$, $\pi$]

```python
x, f, error=np.loadtxt("2.2/2.2.2/Vettore_media.dat", usecols=(0,1,2), unpack='true')

plt.figure(figsize=(9,6))
plt.errorbar(x,f,yerr=error, linewidth=4, zorder=0)
plt.plot(x,pow(x, 0.5), color='r', linewidth=1.2, zorder=10)

plt.legend(['Np^(1/2)', 'Dati'], loc='best')
plt.xlabel('Np')
plt.ylabel('Radice quadrata della distanza quadratica da origine')
plt.grid(True)

plt.show()
```

Come atteso, entrambi i grafici ($\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ ) seguono un andamento diffusivo ($\sqrt{N}$)
