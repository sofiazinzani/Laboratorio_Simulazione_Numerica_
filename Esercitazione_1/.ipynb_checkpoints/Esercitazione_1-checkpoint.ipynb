### ESERCIZIO 1.1

Verifichiamo che il valor medio del nostro generatore di numeri pseudo-casuali sia 1/2 e che la sua varianza sia 1/12. Per farlo estraiamo un set (M) di numeri uniformemente distruibuiti e li dividiamo in N blocchi; ne calcoliamo quindi il valor medio per ogni blocco e successivamente calcoliamo media progressiva e varianza, al crescere dei blocchi.
Notiamo che i risultati ottenuti danno un'incertezza statistica entro 1 \sigma, quindi sono compatibili con i valori attesi.

#### ESERCIZIO 1.1.1

Stima con relativo errore dell integrale $\langle r \rangle = \int_0^1 r dr = 1/2$ in funzione del numero di blocchi

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


N=100 #numero blocchi
M=100000 #numero steps totali

x, f, error=np.loadtxt("1.1/1.1.1/Dati.dat", usecols=(0,1,2), delimiter='\t', unpack='true')

x*=int(M/N)

plt.hlines(
    0.5,
    0, # Start of horizontal line
    x[-1], # End == total number of throws, i.e. input data points.
    colors = "red",
    linestyle = "dotted",
    label = "Expected value"
)

plt.errorbar(x,f,yerr=error, ecolor='c')


plt.xlabel('steps totali') # i punti sono uno ogni L=M/N steps (=uno per blocco)
plt.ylabel('<r>')
plt.grid(True)

plt.show()
```

#### ESERCIZIO 1.1.2

Stima con relativo errore dell integrale $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$ in funzione del numero di blocchi 

```python
N=100
M=100000

x, f, error=np.loadtxt("1.1/1.1.2/Dati.dat", usecols=(0,1,2), delimiter='\t', unpack='true')

x*=int(M/N)

plt.hlines(
    1/12,
    0, # Start of horizontal line
    x[-1], # End == total number of throws, i.e. input data points.
    colors = "red",
    linestyle = "dotted",
    label = "Expected value"
)

plt.errorbar(x,f,yerr=error, ecolor='c')


plt.xlabel('steps totali') # i punti sono uno ogni L=M/N steps (=uno per blocco)
plt.ylabel('<(r-1/2)^2>')
plt.grid(True)

plt.show()
```

Dopo aver aggiunto le distribuizioni di probabilit esponenziale e di Cauchy-Lorentz al codice, le abbiamo utilizzate per ottenere degli istogrammi.
Per ognuna delle 3 distribuzioni sono stati generati 4 istogrammi. Per ogni distribuzione: il primo istogramma  stato ottenuto dalle medie di $4\cdot10^4$ numeri generati secondo la distribuzione desiderata, usando un solo numero, il secondo dalla media usando 2 numeri, il terzo usandone 10 e il quarto 100.

Per N=100  stato fatto anche un fit. 
Le medie ottenute dalle prime due distribuzioni si dispongono su una distribuzione Gaussiana centrata nel valore atteso della distribuzione originaria con varianza ${\sigma^2}/{\sqrt{N}}$ (sempre pi stretta al crescere dei numeri utilizzati per formare una media). Per le distribuzionidi Lorentziane invece si applica la generalizzazione del teorema del limite centrale, che permette di dire che le medie di variabili lorenziane si distrubuiscono lungo una lorentziana a loro volta; in questo caso quindi il fit  stato fatto con una Lorentziana.
In quest'ultimo caso il fit ottenuto non approssima bene la Lorentziana e le incertezze ottenute sui valori del fit sono troppo grandi per avere valore.

### Uniforme

```python
n_bins=100
x, f1, f2, f10, f100=np.loadtxt("1.2/Uniform.dat", usecols=(0,1,2,3,4), delimiter='\t', unpack='true')

fig, (ax1, ax2, ax10, ax100) = plt.subplots(1, 4, figsize=(13,2))

ax=ax1
ax.hist(
    f1, n_bins
);
ax.set_title("N=1")

ax=ax2
ax.hist(
    f2, n_bins
);
ax.set_title("N=2")

ax=ax10
ax.hist(
    f10, n_bins
);
ax.set_title("N=10")


ax=ax100
ax.hist(
    f100, n_bins
);
ax.set_title("N=100")


```

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import matplotlib.mlab as mlab
from scipy.stats import norm

def f(x,A,mu,sigma):  # defining the fitting function
    return A * np.exp(-(x-mu)**2/(2.*sigma**2))


alt, bord, _= plt.hist(f100, bins=n_bins)
plt.clf()




centro= bord[:-1]+np.diff(bord)/2

p_opt, p_cov = curve_fit(f, centro, alt, bounds=(0, [500., 1, 2]))

plt.plot(centro, alt, color='c')



x=centro
y_fit = f(x,p_opt[0],p_opt[1],p_opt[2])

plt.plot(x,y_fit, color='r') # plotting fitted function

print("optimized parameters [a,b,c] =")
print(p_opt)
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(p_cov)))

plt.title('Distribuzione lorenziana. N=100')
plt.legend(['Distribuzione', 'Fit'], loc='best')
plt.xlabel('bin')
plt.ylabel('frequenza')
plt.grid(True)



plt.show()
```

```python
def f(x,A,mu,sigma):  # defining the fitting function
    return A * np.exp(-(x-mu)**2/(2.*sigma**2))


alt, bord, _= plt.hist(f100, bins=n_bins)
plt.clf()




centro= bord[:-1]+np.diff(bord)/2

p_opt, p_cov = curve_fit(f, centro, alt, bounds=(0, [500., 1, 2]))

plt.plot(centro, alt, color='c')



x=centro
y_fit = f(x,p_opt[0],p_opt[1],p_opt[2])

plt.plot(x,y_fit, color='r') # plotting fitted function

print("optimized parameters [a,b,c] =")
print(p_opt)
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(p_cov)))

plt.title('Distribuzione esponenziale. N=100')
plt.legend(['Distribuzione', 'Fit'], loc='best')
plt.xlabel('bin')
plt.ylabel('frequenza')

plt.grid(True)



plt.show()
```

### Lorentziana

```python

n_bins=100
x, f1, f2, f10, f100=np.loadtxt("1.2/Lorentzian.dat", usecols=(0,1,2,3,4), delimiter='\t', unpack='true')

fig, (ax1, ax2, ax10, ax100) = plt.subplots(1, 4, figsize=(13,2))

ax=ax1
ax.hist(
    f1, n_bins, range=(-10,10)
);
ax.set_title("N=1")

ax=ax2
ax.hist(
    f2, n_bins, range=(-10,10)
);
ax.set_title("N=2")


ax=ax10
ax.hist(
    f10, n_bins, range=(-10,10)
);
ax.set_title("N=10")


ax=ax100
ax.hist(
    f100, n_bins, range=(-10,10)
);
ax.set_title("N=100")
```

```python
def f(x,gamma,mu):  # defining the fitting function
    return gamma/(np.pi*((x-mu)**2+gamma**2))


alt, bord, _= plt.hist(f100, bins=n_bins, range=(-10,10))
plt.clf()


centro= bord[:-1]+np.diff(bord)/2

p_opt, p_cov = curve_fit(f, centro, alt)

plt.plot(centro, alt, color='c')


x=centro
y_fit = f(x,p_opt[0],p_opt[1])

plt.plot(x,y_fit, color='r') # plotting fitted function

print("optimized parameters [a,b,c] =")
print(p_opt)
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(p_cov)))

plt.xlabel('bin')
plt.ylabel('frequenza')
plt.title('Distribuzione lorenziana. N=100')
plt.legend(['Distribuzione', 'Fit'], loc='best')
plt.grid(True)



plt.show()
```
