ESERCIZIO 1.1.1

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


N=200
M=100000

x, f, error=np.loadtxt("1.1.1/Prova.dat", usecols=(0,1,2), delimiter='\t', unpack='true')

x*=int(M/N)


plt.errorbar(x,f-0.5,yerr=error)


plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

plt.show()
```
