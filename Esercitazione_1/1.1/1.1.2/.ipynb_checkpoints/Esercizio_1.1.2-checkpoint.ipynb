```python
file=open("Prova.dat", "r")


contenuto=file.read()
righe=contenuto.split("\n")
ave_progr=[]
error_progr=[]
for riga in righe[:-1]:
    elementi=riga.split("\t")
#    print(elementi)
    ave_progr.append(float(elementi[1]))
    error_progr.append(float(elementi[2]))
    
#somma=[elemento[1] for riga in righe for elemento in riga]
#print(colonna)
file.close()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

N=200
M=100000
f=np.asfarray(ave_progr)
x=np.arange(N)
x*=int(M/N)

plt.errorbar(x, f-1/12, yerr=error_progr)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
