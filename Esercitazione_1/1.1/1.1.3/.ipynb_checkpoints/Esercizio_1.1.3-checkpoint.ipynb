```python
file=open("Chi2.dat", "r")


contenuto=file.read()
righe=contenuto.split("\n")

indice=[]
chi2=[]
for riga in righe[:-1]:
    elementi=riga.split("\t")

    indice.append(float(elementi[0]))
    chi2.append(float(elementi[1]))
    

file.close()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


x=np.asfarray(indice)
f=np.asfarray(chi2)

plt.plot(x,f)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
