```python
file=open("Dati.dat", "r")

k=10000
contenuto=file.read()
righe=contenuto.split("\n")

uniform1=[]
esponential1=[]
lorentzian1=[]

uniform2=[]
esponential2=[]
lorentzian2=[]

uniform10=[]
esponential10=[]
lorentzian10=[]

uniform100=[]
esponential100=[]
lorentzian100=[]


for riga in righe[:k]:
    elementi=riga.split("\t")
    print(elementi)
    uniform1.append(float(elementi[1]))
    esponential1.append(float(elementi[2]))
    lorentzian1.append(float(elementi[3]))
    
    
for riga in righe[k:2*k]:
    elementi=riga.split("\t")
    print(elementi)
    uniform2.append(float(elementi[1]))
    esponential2.append(float(elementi[2]))
    lorentzian2.append(float(elementi[3]))
    
for riga in righe[2*k:3*k]:
    elementi=riga.split("\t")
    print(elementi)
    uniform10.append(float(elementi[1]))
    esponential10.append(float(elementi[2]))
    lorentzian10.append(float(elementi[3]))
    
for riga in righe[3*k:4*k]:
    elementi=riga.split("\t")
    print(elementi)
    uniform100.append(float(elementi[1]))
    esponential100.append(float(elementi[2]))
    lorentzian100.append(float(elementi[3]))
    
#somma=[elemento[1] for riga in righe for elemento in riga]
#print(colonna)
file.close()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

N=10000
f1=np.asfarray(uniform1)
f2=np.asfarray(uniform2)
f10=np.asfarray(uniform10)
f100=np.asfarray(uniform100)
x=np.arange(N)
n_bins=50


fig, (ax1, ax2, ax10, ax100) = plt.subplots(1, 4, figsize=(13,2))

ax=ax1
ax.hist(
    f1, n_bins
);

ax=ax2
ax.hist(
    f2, n_bins
);


ax=ax10
ax.hist(
    f10, n_bins
);


ax=ax100
ax.hist(
    f100, n_bins
);
```

```python
from scipy.stats import norm
import matplotlib.mlab as mlab


N=10000
f1=np.asfarray(uniform1)
f2=np.asfarray(uniform2)
f10=np.asfarray(uniform10)
f100=np.asfarray(uniform100)
x=np.arange(N)
n_bins=50


fig, (ax1, ax2, ax10, ax100) = plt.subplots(1, 4, figsize=(13,2))

ax=ax1
ax.hist(
    f1, n_bins
);

ax=ax2
ax.hist(
    f2, n_bins
);


ax=ax10
ax.hist(
    f10, n_bins
);


ax=ax100


# best fit of data
(mu, sigma) = norm.fit(uniform100)

# the histogram of the data

ax.hist(
    f100, n_bins, facecolor='green'
);

# add a 'best fit' line
y = mlab.normpdf(n_bins, mu, sigma)
l = plt.plot(n_bins, y, 'r--', linewidth=2)

#plot
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))
plt.grid(True)
_
plt.show()
```
