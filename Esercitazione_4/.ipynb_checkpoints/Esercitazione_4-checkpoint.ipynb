## Esercizio 4.1

Nella cartella Termalizzazione sono presenti gli esiti di più simulazioni (qui 4) consecutive, senza data blocking. I risultati di questo tipo di simulazione sono riportati nel primo grafico. 
Le config.final e config.prec ottenute tramite termalizzazione sono riportate nella cartella Data Blocking come invece config0.final e config0.prec, mantenute per ricominciare la simulazione da un punto in cui sia già stata termalizzata (usandole per sovrascrivere config.final e config.prec). 
Nel momento in cui poi viene effettuata una simulazione con data blocking queste verranno sovrascritte. 

Studiando la temperatura, notiamo che nelle prime simulazioni abbiamo dei salti netti, mentre successivamente il sistema tende a stabilizzarsi: le correzioni iniziali sono infatti più grandi.
Le unità di misura sono quelle ridotte.

gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)

solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.0$)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure(figsize=(12, 9))

plt.subplot(3,1,1)
file=open("Gas/Termalizzazione/output_temp.dat", "r")
f=np.loadtxt("Gas/Termalizzazione/output_temp.dat", usecols=(0))
plt.plot(range(len(f)),f, color = 'm')
plt.xlabel('N_step')
plt.ylabel('T')
plt.grid(True)


plt.subplot(3,1,2)
file=open("Liquido/Termalizzazione/output_temp.dat", "r")
f=np.loadtxt("Liquido/Termalizzazione/output_temp.dat", usecols=(0))
plt.plot(range(len(f)),f, color = 'c')
plt.xlabel('N_step')
plt.ylabel('T')
plt.grid(True)


plt.subplot(3,1,3)
file=open("Solido/Termalizzazione/output_temp.dat", "r")
f=np.loadtxt("Solido/Termalizzazione/output_temp.dat", usecols=(0))
plt.plot(range(len(f)),f, color = 'b')
plt.xlabel('N_step')
plt.ylabel('T')
plt.grid(True)




plt.tight_layout()
plt.show()
```

## Esercizio 4.2-4.3

E' stato quindi eseguito data blocking sul sistema termalizzato, riportando così i valori medi per unità di particella e le sue incertezza per: temperatura, energia cinetica, energia potenziale e energia totale.
Esso è stato eseguito per tutti e tre i sistemi: gas, liquido e solido, con i parametri iniziali definiti precedentemente.

```python
sigma=0.34e-9
e_su_kb=120
m=39.948

kb=1.381*pow(10,-23)
e=e_su_kb*kb

integration_step=0.0005


time=np.sqrt(m*sigma**2/e)
```

### Temperatura

```python


fig=plt.figure(figsize=(12, 9))

plt.subplot(3,1,1)
x, f, error = np.loadtxt("Gas/DataBlocking/Risultati/ave_temp.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
#plt.errorbar(x*integration_step*time,f*e_su_kb,yerr=error*e_su_kb, elinewidth=1, capsize=1, ms=3)
plt.errorbar(x,f*e_su_kb,yerr=error*e_su_kb, elinewidth=1, capsize=1, ms=3, color = 'm')
plt.xlabel('N_blk')
plt.ylabel("T [K]")
plt.title("Data Blocking: Temperature")
plt.grid(True)



plt.subplot(3,1,2)
x, f, error = np.loadtxt("Liquido/DataBlocking/Risultati/ave_temp.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f*e_su_kb,yerr=error*e_su_kb, elinewidth=1, capsize=1, ms=3, color = 'c')
plt.xlabel('N_blk')
plt.ylabel("T [K]")
plt.title("Data Blocking: Temperature")
plt.grid(True)


plt.subplot(3,1,3)
x, f, error = np.loadtxt("Solido/DataBlocking/Risultati/ave_temp.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f*e_su_kb,yerr=error*e_su_kb, elinewidth=1, capsize=1, ms=3, color = 'b')
plt.xlabel('N_blk')
plt.ylabel("T [K]")
plt.title("Data Blocking: Temperature")
plt.grid(True)


plt.tight_layout()
plt.show()
```
