# Esercitazione 5

I dati riportati mostrano l'andamento del valor medio della distanza dall'origine delle posizioni generate in frazioni del raggio di Bohr $a_0$ al variare del numero di blocchi, utilizzati per calcolare la media.
I valori di aspettazione per il raggio sono:
$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$
</span>


Osserviamo l'utilizzo di una distribuzione di probabilità gaussiana centrata nella vecchia posizione con ampiezza delta fornisce risultati analoghi a quelli ottenuti con una probabilità di transizione uniforme.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


fig=plt.figure(figsize=(12, 4))

plt.subplot(1,2,1)
x, f, error=np.loadtxt("5.1/Risultati0_Unif.dat", usecols=(0,1,2), delimiter='\t', unpack='true')
plt.errorbar(x,f,yerr=error, color = 'm')

plt.hlines(
    3/2,
    0, # Start of horizontal line
    x[-1], # End == total number of throws, i.e. input data points.
    colors = "red",
    linestyle = "dotted",
    label = "Expected value"
)

plt.title('Distribuzione Uniforme')
plt.legend(['Valore atteso', 'Integrale calcolato'], loc='best')
plt.xlabel('Nblocchi')
plt.ylabel(r"$\left\langle r \right\rangle_{\Psi_{1,0,0}}/a_0$")
plt.grid(True)








plt.subplot(1,2,2)
x, f, error=np.loadtxt("5.1/Risultati1_Unif.dat", usecols=(0,1,2), delimiter='\t', unpack='true')
plt.errorbar(x,f,yerr=error, color = 'b')


plt.hlines(
    5,
    0, # Start of horizontal line
    x[-1], # End == total number of throws, i.e. input data points.
    colors = "red",
    linestyle = "dotted",
    label = "Expected value"
)

plt.title('Distribuzione Uniforme')
plt.legend(['Valore atteso', 'Integrale calcolato'], loc='best')
plt.xlabel('Nblocchi')
plt.ylabel(r"$\left\langle r \right\rangle_{\Psi_{1,0,0}}/a_0$")

plt.grid(True)




plt.tight_layout()
plt.show()
```

### Rappresentazioni 3D

```python
from mpl_toolkits.mplot3d import Axes3D
plt.rc('axes', titlesize=5)

x, y, z = np.loadtxt("5.1/Posizioni.dat", usecols=(0,1,2), delimiter='	', unpack='true',max_rows=10000)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x, y, z, c=x, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10,                                                                                                                                                                                                  5)
plt.show()
```
