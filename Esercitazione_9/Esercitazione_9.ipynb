# Esercizio 9.1

Per generare i dati: eseguire il comando *make*, seguito da *./CityGenerator.exe* e *./main.exe*.

Nel file *CityGenerator.cpp* si creano le posizioni delle città da raggiungere e vengono salvate in *CityPosition.dat*. Le città sono disposte su una circonferenza o in un quadrato a seconda che nel file *main.h* venga passato *true* o *false* alla variabile booleana *Circonferenza*.
Nel *main.cpp* invece, date le posizioni di NCity città, si determina il percorso più breve per visitarle tutte una sola volta e tornare ritornando al punto di partenza. 
Questo viene fatto sfruttando un algoritmo genetico che genera *Npopulation* popolazioni di *Nelement* percorsi l'una. Seleziona per ogni popolazione creata il percorso migliore e salva la sua lunghezza e la lunghezza media della metà migliore di popolazione nel file *Best.dat*.

Di seguito sono riportati i grafici della lunghezza del percorso più breve e la media delle lunghezze della metà migliore della popolazione, in funzione del numero di popolazioni generate. Infine è riportato il miglior percorso ottenuto.

## Circonferenza

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import string
```

```python
fig=plt.figure(figsize=(10, 5))

npop, best, halfbest = np.loadtxt("9.1/Circonferenza/Best.dat", usecols=(0,1,2), delimiter='	', unpack='true')

plt.subplot(2, 1, 1)
plt.plot(npop,best, '.-', color = 'm')
plt.xlabel('Numers of population')
plt.ylabel('Length')
plt.title('Best Trip Length')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(npop,halfbest, '.-', color = 'm')
plt.xlabel('Numers of population')
plt.ylabel('Length')
plt.title('Trip Length of half best population average')
plt.grid(True)

plt.tight_layout()
#plt.savefig("./pictures/BestTripsLength.png")

plt.show()
```

```python
fig=plt.figure(figsize=(10, 5))

i, x, y = np.loadtxt("9.1/Circonferenza/BestTrip.dat", usecols=(0,1,2), delimiter='	', unpack='true')

title = str(np.loadtxt("9.1/Circonferenza/BestLength.dat"))
plt.plot(x,y,'.-b', color = 'm')
plt.title('Length: '+title)
plt.grid()

plt.tight_layout()

plt.show()
```

## Quadrato

```python
fig=plt.figure(figsize=(10, 5))

npop, best, halfbest = np.loadtxt("9.1/Best.dat", usecols=(0,1,2), delimiter='	', unpack='true')

plt.subplot(2, 1, 1)
plt.plot(npop,best, '.-', color = 'g')
plt.xlabel('Numers of population')
plt.ylabel('Length')
plt.title('Best Trip Length')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(npop,halfbest, '.-', color = 'g')
plt.xlabel('Numers of population')
plt.ylabel('Length')
plt.title('Trip Length of half best population average')
plt.grid(True)

plt.tight_layout()

plt.show()
```
